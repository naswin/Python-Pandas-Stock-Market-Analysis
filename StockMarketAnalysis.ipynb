####     1.Here all the functions necessary to prepare the data, and perform analysis are defined

```python
#Function to get top N Market Cap Companies from a given Sector
def getTopMC(df,N,sectorName):
    TopMCSymbolList = []
    TopMCSymbolList = df[(df.Sector==sectorName)].sort(columns='Market Cap',ascending=False).head(N).Symbol.tolist()
    return TopMCSymbolList 


#function to get the daily data from yahoo for each of the ticker symbols for last 10 years
def getYahooStockData(StockList,numOfYears):
    end = dt.datetime.now()
    start = dt.datetime(end.year-numOfYears,end.month,end.day)
    stockInfo_dict = dict()
    for item in StockList:
        stockInfo_dict[item] = DataReader(item,'yahoo',start,end)
    return stockInfo_dict

#clean the dataframes to only have the Adj Close column
def cleanDf(stockDict):
    
    for key in stockDict:
        stockDict[key] = stockDict[key]['Adj Close']
    return stockDict

#function to compute the daily returns for the sector stocks
def getDailyReturns(stockDict):
    for key in stockDict:
        stockDict[key]['Daily Return'] = np.log(stockDict[key]['Adj Close']/stockDict[key]['Adj Close'].shift(1))
    return stockDict

#function to aggregate the daily returns to a monthly level
def aggregateDailyReturns(stockDict):
    for key in stockDict:
        stockDict[key]['Monthly Return'] = stockDict[key]['Daily Return'].asfreq('EOM',method='ffill')
    return stockDict

#function to resize the dataframe to the size of S&P500 dataframe
def resizeDataFrames(stockDict):
    for key in stockDict:
        if len(stockDict[key].index) != len(stockDict['^GSPC'].index):
            stockDict[key] = stockDict[key].reindex(index=stockDict['^GSPC'].index)
    return stockDict
    

#main function to call the above data preparation functions
def main(sectorName,numberOfStocks,numberOfYears):
    #get the desired list of stocks for a given tech sector
    TopMCSymbolList = getTopMC(sp500StockDetail_df,numberOfStocks,sectorName)
    #append the s&p500 index ticker to the ticker list
    TopMCSymbolList.append('^GSPC')
    
    #get the daily stock price info for a desired number of years
    stockInfo_dict = getYahooStockData(TopMCSymbolList,numberOfYears)

    #getDailyReturn data frame
    stockDict_DailyReturns = getDailyReturns(stockInfo_dict)

    #aggregate the daily returns to a monthly level
    stockDict_MonthlyReturns = aggregateDailyReturns(stockInfo_dict)

    #resize the Daily and Monthly returns dataframes
    stockDict_DailyReturns_RS = resizeDataFrames(stockInfo_dict)
    
    return stockInfo_dict
```

#### 9. Testing for Normality of our time series data sets using the Shapiro Test

```python
def shapiroTest(df):
    shapiro_dict = {}
    for col in df.columns:
        shapiro_dict[col] = sts.shapiro(df[col].dropna())
    return shapiro_dict
#run the shapiro test on the merged_df dataframe containing our time series data
shapiroDict = shapiroTest(merged_df)

#conver the returned dictionary into a dataframe
shapiro_df = pd.DataFrame(shapiroDict)

#retain only the first index which contains the p-value for testing normality
shapiro_df = shapiro_df.drop(shapiro_df.index[[0]])

#return a results dataframe with True/False values of whether the p-value is greater than 0.05
shapiro_results = shapiro_df > 0.05

#transpose the results dataframe to have the p-value result as a column instead of index
shapiro_results = shapiro_results.T
shapiro_results.columns = ['Is p-value greater than 0.05']

shapiro_results
```

###### With the above results, that the p-value for none our timeseries distrisbutions is above 0.05, we reject the Null Hypothesis that any of our timeseries distributions follow a normal distribution
